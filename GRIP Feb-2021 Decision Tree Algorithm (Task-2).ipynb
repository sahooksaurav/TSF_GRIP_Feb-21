#### 7) Data Preprocessing

```python
# Removing the unwanted columns from the data
data = df.iloc[:,1:5]
```

```python
target = df['Species']
```

#### 8) Performing Train Test Split

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y = train_test_split(data,target, random_state = 101)
```

```python
# Percentage of each species in training data
print('Percentage of each species in training data')
train_y.value_counts()/len(train_y)*100
```

```python
#Percentage of each species in test data
print('Percentage of each species in test data')
test_y.value_counts()/len(test_y)*100
```

#### 9) Training the model

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(train_x, train_y)
```

#### 8) Visualizing the tree

```python
c = list(data.columns)
t = df['Species'].unique()

plt.figure(figsize=(25,25))
plot_tree(clf, feature_names = c, class_names = t, filled = True)
plt.show()
```

```python
#predicted values
predicted_y = clf.predict(test_x)
predicted_y
```

#### 11) Evaluating Metrics

```python
import sklearn.metrics as sm
```

```python
print('Accuracy score of model is: {0}'.format(sm.accuracy_score(test_y, predicted_y)))
```

```python
sm.confusion_matrix(test_y, predicted_y)
```

#### 12) Training the model with Entropy criterion

```python
clf2 = DecisionTreeClassifier(criterion = 'entropy')
```

```python
clf2.fit(train_x,train_y)
```

#### 13) Visualizing the tree

```python
c = list(data.columns)
t = df['Species'].unique()

plt.figure(figsize=(25,25))
plot_tree(clf2, feature_names = c, class_names = t, filled = True)
plt.show()
```

```python
pred_y = clf2.predict(test_x)
pred_y
```

#### 14) Evaluation Metrics

```python
print('Accuracy score of model is: {0}'.format(sm.accuracy_score(test_y, pred_y)))
```

```python
sm.confusion_matrix(test_y, pred_y )
```
